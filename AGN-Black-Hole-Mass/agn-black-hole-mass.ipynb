```python
# Importing modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing dataset
data = pd.read_csv(r'https://raw.githubusercontent.com/SuperbSaud/astro-data-science/main/AGN-Black-Hole-Mass/blackhole.csv')
```

```python
print(data.shape)
print(data.columns)
```

```python
# sort data frame
data.sort_values(["log M_BH"], 
                    axis=0,
                    ascending=[True], 
                    inplace=True)
```

```python
BH_mass = data['log M_BH']
BH_redshift = data['Redshift']
```

```python
# Scatter plot for Mass and Redshift
plt.figure(figsize=(20,10))
plt.scatter(x=BH_mass, y=BH_redshift, c='r')
plt.title('Mass & Redshift Scatter Plot')
plt.xlabel('Black hole mass')
plt.xticks(rotation=90)
plt.ylabel('Redshift')
plt.show()
```
